```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Constants
k = 1.38e-23  # Boltzmann's constant (J/K)
q = 1.6e-19   # Electron charge (C)

#  Diode Parameter
# Let model  be a typical silicon diode
Is = 1e-12    # Reverse saturation current (A)
n = 1.2         # Ideality factor
T_celsius = 27  # Temperature in Celsius
```

```python
# Convert temperature to Kelvin
T_kelvin = T_celsius + 273.15

# Calculate thermal voltage (Vt)
Vt = (k * T_kelvin) / q

print(f"Temperature: {T_kelvin:}K")
print(f"Thermal Voltage (Vt): {Vt:} V")
```

```python
Vd = np.linspace(-1, 0.7, 1000)
Id = Is * (np.exp(Vd / (n * Vt)) - 1)
```

```python
plt.plot(Vd, Id, color='b') 
plt.title('PN Junction Diode I-V Characteristic', fontsize=16)
plt.xlabel('Diode Voltage (Vd) [V]', fontsize=12)
plt.ylabel('Diode Current (Id) [A]', fontsize=12)
plt.grid(True)
plt.show()
```

```python
# **** Temperature Analysis ****

# list of temp to loop 
temps_celsius = [0, 25, 50, 100]

# Loop through each temperature
for T_c in temps_celsius:
    T_k = T_c + 273.15 # Kelvin conversion
    Vt_loop = (k * T_k) / q 
    Id_loop = Is * (np.exp(Vd / (n * Vt_loop)) - 1)   
    plt.plot(Vd, Id_loop, linewidth=2, label=f'{T_c}C')

# FINAL PLOTTING
plt.yscale('log')
plt.title('I-V Curve Dependence on Temperature', fontsize=16)
plt.xlabel('Diode Voltage (Vd) [V]', fontsize=12)
plt.ylabel('Diode Current (Id) [A]', fontsize=12)
plt.grid(True)
      
plt.legend()          
plt.show()

```
